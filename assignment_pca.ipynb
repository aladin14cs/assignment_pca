```python
import numpy as np 
import pandas as pd
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
wine=pd.read_csv('wine.csv')
```

```python
wine.head()
```

```python
wine.shape
```

```python
wine_data=wine.iloc[:,1:]
wine_data.values
```

```python
WINE=wine_data.values
```

```python
WINE
```

```python
wine_norm=scale(WINE)
```

```python
wine_norm
```

```python
pca=PCA()
pca_wine=pca.fit_transform(wine_norm)
pca_wine
```

```python
pca=PCA(n_components=13)
pca_wine=pca.fit_transform(wine_norm)
pca_wine
```

```python
pca.components_
```

```python
wine_var=pca.explained_variance_ratio_
wine_var
```

```python
wine_var1=np.cumsum(np.round(wine_var,decimals=4)*100)
wine_var1
```

```python
plt.plot(wine_var1,color="red")
```

```python
pca_wine[:,0:1]
```

```python
finalDf = pd.concat([pd.DataFrame(pca_wine[:,0:2],columns=['pc1','pc2']), wine[['Type']]], axis = 1)
finalDf
```

```python
#kmeans clustering

from sklearn.cluster import KMeans

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(pca_wine)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
clusters_new = KMeans(4, random_state=42)
clusters_new.fit(pca_wine)
```

```python
clusters_new.labels_
```

```python
wine['clusterid_new'] = clusters_new.labels_
```

```python
clusters_new.cluster_centers_
```

```python
wine.groupby('clusterid_new').agg(['mean']).reset_index()
wine
```
